```python
#importing libraires
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#importing datasets
dataset=pd.read_csv('z-csv/Position_Salaries.csv')
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
#excaptionally we have very less value in dataset so we directly tarin our model without splitting the data
#Training Linear regression model on the dataset
from sklearn.linear_model import LinearRegression
linear_regressor=LinearRegression()
linear_regressor.fit(x,y)
```

```python
#Training polynomial regression model on the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
linear_regressor_2=LinearRegression()
linear_regressor_2.fit(x_poly,y)
```

```python
#visualising linear regression result
plt.scatter(x,y,color='red')
plt.plot(x,linear_regressor.predict(x),color='blue')
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
#visualising Polynomial linear regression result
plt.scatter(x,y,color='red')
plt.plot(x,linear_regressor_2.predict(poly_reg.fit_transform(x)),color='blue')
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
#visualising Polynomial linear regression result(for higher resolution and smoother curve) 
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape(len(x_grid),1)

plt.scatter(x,y,color='red')
plt.plot(x_grid,linear_regressor_2.predict(poly_reg.fit_transform(x_grid)),color='blue')
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
#Predicting new result with Polynomial Regression
print(linear_regressor_2.predict(poly_reg.fit_transform([[6.5]])))
```

```python
#Predicting new result with Linear Regression
print(linear_regressor.predict([[6.5]]))
```
